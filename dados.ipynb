```python
import pandas as pd

url = "https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/qus/ultimas-4-semanas-gasolina-etanol.csv"

df = pd.read_csv(url, sep=';', encoding='latin1', decimal=',')

display(df)
```

```python
gasolina_df = df[df["Estado - Sigla"] == "DF"]#filtra o DF

gasolina_df = gasolina_df[gasolina_df['Produto'].str.contains('GASOLINA', na=False)]#busca tudo que tem gasolina

gasolina_df['Data da Coleta'] = pd.to_datetime(gasolina_df['Data da Coleta'], format='%d/%m/%Y')#formata data

gasolina_df = gasolina_df.drop('Valor de Compra', axis=1)#apaga oq nao quero
gasolina_df = gasolina_df.drop('Complemento', axis=1)
gasolina_df = gasolina_df.dropna(subset=['Valor de Venda', 'Data da Coleta'])#remove nulos por linha

display(gasolina_df)
```

```python

resultado_agrupado = gasolina_df.groupby("Data da Coleta")["Valor de Venda"].agg(
    Preco_Medio='mean',
    Preco_Minimo='min',
    Preco_Maximo='max',
    Qtd_Postos='count'
).reset_index() 


display(resultado_agrupado)
```

```python
resultado_agrupado = resultado_agrupado.rename(columns={
    'Data da Coleta': 'data_coleta',
    'Preco_Medio': 'preco_medio',
    'Preco_Minimo': 'preco_minimo',
    'Preco_Maximo': 'preco_maximo',
    'Qtd_Postos': 'qtd_postos'
})
```

```python
import sqlite3


con = sqlite3.connect("gasolina.db")
cursor = con.cursor()


cursor.execute("""
CREATE TABLE IF NOT EXISTS preco (
    data_coleta DATE PRIMARY KEY,
    preco_medio REAL,
    preco_minimo REAL,
    preco_maximo REAL,
    qtd_postos INTEGER
)
""")
con.commit()
cursor.close()


resultado_agrupado.to_sql(
    "preco",
    con,
    if_exists="append",
    index=False
)


con.close()

print("Tabela criada com tipo DATE e dados salvos com sucesso!")
```
